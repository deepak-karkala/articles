```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
W = 80;
G = 0 #0.2 * 100;
Crr = 0.005;
Cd = 0.63;
A = 0.509;
Rho = 1.22601;

power = []
velocity = []
fra = []
fga = []
fda = []
for vkh in np.arange(0, 50, 0.1):
    v = vkh * 1000/3600
    fr = 9.8067 * W * Crr * np.cos(np.arctan(G/100))
    fg = 9.8067 * W * np.sin(np.arctan(G/100))
    fd = 0.5*Cd*A*Rho*v*v
    p = ((9.8067 * W * ( np.sin(np.arctan(G/100)) + Crr * np.cos(np.arctan(G/100)) )) + (0.5*Cd*A*Rho*v*v))*v;
    #p = (0.5*Cd*A*Rho*v*v)*v;
    power.append(p)
    velocity.append(vkh)
    fra.append(fr)
    fga.append(fg)
    fda.append(fd)
```

```python
np.arange(0,10,0.1)
```

```python
plt.figure()
plt.plot(velocity, power)
```

```python
plt.figure()
plt.plot(velocity, fga)
plt.plot(velocity, fra)
plt.plot(velocity, fda)
```

```python
df = pd.DataFrame()
df["bins"] = velocity
df["power"] = power
df.to_csv("../code/data/power_velocity.csv", index=False)
df
```
