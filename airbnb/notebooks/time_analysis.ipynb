```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
pd.set_option("display.max_columns", 100)
import os
import geojson
from geojson import Feature, FeatureCollection
```

```python
basepath = "/Volumes/Seagate Backup Plus Drive/data_journalism/airbnb/raw_data/cities/"
```

```python
analysis_cities = ["New York City", "Paris", "Los Angeles", "London", "Berlin", "Barcelona", "Amsterdam", "San Francisco"]
```

```python
dfl = pd.DataFrame()
idx = 0

for city in os.listdir(basepath):
    
    if city in analysis_cities:
        print(city)
        for date in os.listdir(basepath+city):
            filepath = basepath + city + "/" + date
            df = pd.read_csv(filepath + "/listings.csv")
            
            dfc = pd.DataFrame()
            dfc.at[idx, "city"] = city
            dfc.at[idx,"date"] = date
            dfc.at[idx, "num_listings"] = len(df)
            
            idx += 1
            dfl = dfl.append(dfc)
dfl.sort_values("date", inplace=True)
```

```python
dfl.sort_values("date", inplace=True)
dfl.to_csv("../code/data/num_listings_over_time.csv", index=False)
```

```python
plt.figure(figsize=(18,6))
city = "San Francisco"
dfc = dfl[dfl["city"]==city].copy()
dfc.sort_values("date", inplace=True)
plt.plot(dfc["date"], dfc["num_listings"])
```
