```python
url = "http://insideairbnb.com/get-the-data.html"
res = requests.get(url)
soup = BeautifulSoup(res.content,'lxml')
```

### List of cities

```python
import os
cities = []
r = soup.find_all('td')[1::4]
for i in range(len(r)):
    city = r[i].text
    if city not in cities:
        cities.append(city)
        #os.mkdir("/Volumes/ext64/backup/aim/journalism/articles/airbnb/data/cities/"+city)
        #os.mkdir("/Users/nesara/Documents/etc/airbnb/data/cities/"+city)
```

```python
r = soup.find_all('td')
citiesDict = {}

for i in range(len(r)):
    if r[i].text in cities:
        city = r[i].text
        if city not in citiesDict:
            print(city)
            citiesDict[city] = 1
        else:
            citiesDict[city] += 1
    
    if r[i].find_all('a') and citiesDict[city]<=7:
        text = r[i].find_all('a')[0].text
        if text=="listings.csv" or text=="neighbourhoods.csv" or text=="neighbourhoods.geojson":
            url = r[i].find_all('a')[0]["href"]
            print(url)
```

```python
basepath = "/Volumes/Seagate Backup Plus Drive/data_journalism/airbnb/cities/"

r = soup.find_all('td')
citiesDict = {}
count = 1;
excluded_cities = ["Malaga"]
missing_files_url = []

#for i in range(len(r)):
for i in range(11467, len(r)):
    
    if ((r[i].text not in excluded_cities) and (r[i].text in cities)):
        city = r[i].text
        if city not in citiesDict:
            print(city)
            citiesDict[city] = 1
            if not os.path.exists(basepath+city):
                os.mkdir(basepath+city)
        else:
            citiesDict[city] += 1
    
    if ((city not in excluded_cities) and r[i].find_all('a')): #and citiesDict[city]<=7:
        text = r[i].find_all('a')[0].text
        if text=="listings.csv" or text=="neighbourhoods.csv" or text=="neighbourhoods.geojson":
            url = r[i].find_all('a')[0]["href"]
            date = url.split("/")[6]
            filepath = basepath+city+"/"+date
            if not os.path.exists(filepath):
                os.makedirs(filepath)   
            #print(url)
            try:
                wget.download(url, filepath)
            except:
                print(url)
                missing_files_url.append(url)
            count += 1
```
