### Ecological variant

```python
players = [axl.Cooperator(), axl.Defector(),
           axl.TitForTat(), axl.Grudger(),
           axl.Random()]
tournament = axl.Tournament(players)
results = tournament.play()
eco = axl.Ecosystem(results)
eco.reproduce(100)
```

```python
players
```

```python
eco.population_sizes
```

```python
dfe = pd.DataFrame(eco.population_sizes)
scale_factor = 1000
dfe.columns = ["Cooperator", "Defector", "Tit For Tat", "Grudger", "Random"]
for col in dfe.columns:
    dfe[col] *= scale_factor
    dfe[col] = np.round(dfe[col],0)
    dfe[col] = dfe[col].astype(int)
```

```python
dfe.reset_index(inplace=True)
dfe.rename(columns={"index":"round"}, inplace=True)
dfe.to_csv("../code/data/eco_results.csv", index=False)
dfe
```
