## Process ball-by-ball for all matches 

```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""


def ball_by_ball_details(url):
    
    res = requests.get(url)
    soup1 = BeautifulSoup(res.content,'lxml')
    matches = soup1.find_all('a', {'class': 'text-hvr-underline', 'href': True})
    venues = soup1.find_all("div", {"class":"text-gray"})
    matchid_offset = 1

    # Winner details
    winners_list = soup1.find_all("a", {"class":"cb-text-link"})
    winners_details = []
    dl_applied_list = []
    for i in range(len(winners_list)):
        win_details = winners_list[i].text
        if win_details.endswith("(DLS method)"):
            win_details = win_details[:-13]
            dl_applied_list.append(1)
        else:
            dl_applied_list.append(0)
        win_details = win_details.split(" won by ")
        winners_details.append(win_details)
        
    if url=="https://www.cricbuzz.com/cricket-series/2717/caribbean-premier-league-2018/matches":
        winners_details[24] = ['St Kitts and Nevis Patriots', '7 wkts']        
    
    dfballbyball = pd.DataFrame()

    # Iterate over all matches
    matches = matches[:-2]
    
    for i in range(len(matches)):

        print(i)
        dfmatch = pd.DataFrame()

        match_href = "http://www.cricbuzz.com" + matches[i]['href']
        venue = venues[i*2+1]
        city = venue.text.split(", ")[1]
        venue = venue.text
        match_id = matchid_offset + i
        season = 2018

        team1 = matches[i].find_all('span')[0].text.split(" vs ")[0]
        team2 = matches[i].find_all('span')[0].text.split(" vs ")[1].split(", ")[0]

        winner = winners_details[i][0]
        is_super_over = 0
        dl_applied = dl_applied_list[i]
        result = "normal"


        ## Details not available through scraping
        date = np.nan
        toss_winner = np.nan
        toss_decision = np.nan
        player_of_match = np.nan
        umpire1 = np.nan
        umpire2 = np.nan
        umpire3 = np.nan


        if winner == team1:
            loser = team2
        else:
            loser = team1

        if (winners_details[i] != ["No result"]) and (winners_details[i] != ["Match tied"]) and (
            winners_details[i] != ["Match abandoned without a ball being bowled(no toss)"]) and (
            winners_details[i] != ["Match abandoned without a ball being bowled (no toss)"]) and (
            winners_details[i] != ["Match abandoned without a ball being bowled (no toss) due to wet outfield"]) and (
            winners_details[i] != ["Match abandoned without toss due to rain"]) and (
            winners_details[i] != ["Match abandoned due to rain (without a ball being bowled)"]) and (
            winners_details[i] != ["Match tied (Islamabad United won the one-over eliminator)"]) and (
            winners_details[i] != ["Match tied (Lahore Qalandars won the super over) "]) and (
            winners_details[i] != ["Match abandoned without a ball being bowled"]) and (
            winners_details[i] != ["Match abandoned without toss due to wet outfield"]) and (
            winners_details[i] != ["Match abandoned without toss due to rain"]) and (
            winners_details[i] != ["Match rescheduled to 01/09/2018 due to bad outfield"]) and (
            winners_details[i] != ["Match tied (Bellary Tuskers won the one-over eliminator)"]) and (
            winners_details[i] != ["Match tied ( Lyca Kovai Kings won super over)"]) and (
            winners_details[i] != ["Match abandoned (postponed) without a ball being bowled(no toss)"]) and (
            winners_details[i] != ["Match tied - Punjab won the one-over eliminator"]):
            if "run" in winners_details[i][1]:
            #if winners_details[i][1].endswith=="run" or winners_details[i][1].endswith=="runs":
                win_by_runs = int(winners_details[i][1].split(" ")[0])
                win_by_wickets = 0       
                innings1_batting_team = winner
                innings1_bowling_team = loser
                innings2_batting_team = loser
                innings2_bowling_team = winner
            elif "wkt" in winners_details[i][1]:
                win_by_wickets = int(winners_details[i][1].split(" ")[0])
                win_by_runs = 0
                innings1_batting_team = loser
                innings1_bowling_team = winner
                innings2_batting_team = winner
                innings2_bowling_team = loser
            else:
                print("Error in win detail !!")


            res = requests.get(match_href)
            soup2 = BeautifulSoup(res.content,'lxml')
            ball_id = soup2.find_all("span",{"class":"text-bold","class":"cb-col"})[::-1]
            ball_details = soup2.find_all("p",{"class":"cb-col-90"})[::-1]

            if len(ball_id) != len(ball_details):
                print("Error !!")
                break
            else:
                inning = 1
                last_ball_batsman = ""
                non_striker = ""

                for j in range(len(ball_id)):

                    over = int(ball_id[j].text.split(".")[0]) + 1
                    ball = int(ball_id[j].text.split(".")[1])

                    ## Set innings number
                    if (j>20 and over==1 and ball==1):
                        inning = 2
                        last_ball_batsman = ""
                        non_striker = ""

                    if inning==1:
                        batting_team = innings1_batting_team
                        bowling_team = innings1_bowling_team
                    else:
                        batting_team = innings2_batting_team
                        bowling_team = innings2_bowling_team

                    bowler = ball_details[j].text.split(", ")[0].split(" to ")[0]
                    batsman = ball_details[j].text.split(", ")[0].split(" to ")[1]

                    if batsman != last_ball_batsman:
                        non_striker = last_ball_batsman
                        last_ball_batsman = batsman

                    # Check different scenarios for each ball
                    ball_event = ball_details[j].text.split(", ")[1]

                    wide_runs = 0
                    bye_runs = 0
                    legbye_runs = 0
                    noball_runs = 0
                    penalty_runs = 0
                    batsman_runs = 0
                    extra_runs = 0
                    total_runs = 0
                    player_dismissed = np.nan
                    dismissal_kind = np.nan
                    fielder = np.nan

                    if ball_event == "wide":
                        wide_runs = 1
                        total_runs = 1
                    elif ball_event == "no ball":
                        noball_runs = 1
                        batsman_runs = 0
                        if len(ball_details[j].text.split(", "))>2 and "run" in ball_details[j].text.split(", ")[2]:
                            try:
                                batsman_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                            except:
                                batsman_runs = 0
                        total_runs = noball_runs + batsman_runs                
                    elif ball_event=="leg byes" or ball_event=="Leg byes":
                        if "run" in ball_details[j].text.split(", ")[2]:
                            legbye_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                            total_runs = legbye_runs
                    elif ball_event == "byes":
                        if "run" in ball_details[j].text.split(", ")[2]:
                            bye_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                            total_runs = bye_runs
                    elif "wides" in ball_event:
                        extra_runs = ball_event.split(" ")[0]
                        total_runs = extra_runs
                    elif "no run" in ball_event:
                        batsman_runs = 0
                        total_runs = 0
                    elif "1 run" in ball_event: # == "1 run" or ball_event=="1 run," or ball_event:
                        batsman_runs = 1
                        total_runs = 1
                    elif ball_event == "2 runs":
                        batsman_runs = 2
                        total_runs = 2
                    elif ball_event == "3 runs":
                        batsman_runs = 3
                        total_runs = 3
                    elif ball_event == "FOUR" or ball_event == "4 runs":
                        batsman_runs = 4
                        total_runs = 4
                    elif ball_event == "5 runs":
                        batsman_runs = 5
                        total_runs = 5
                    elif ball_event == "SIX":
                        batsman_runs = 6
                        total_runs = 6
                    elif ball_event.split(" ")[0] == "out":
                        if ball_event.split(" ")[1] == "Lbw!!":
                            player_dismissed = batsman
                            dismissal_kind = "lbw"
                            last_ball_batsman = non_striker
                        elif ball_event.split(" ")[1] == "Bowled!!":
                            player_dismissed = batsman
                            dismissal_kind = "bowled"
                            last_ball_batsman = non_striker
                        elif ball_event.split(" ")[1] == "Caught":
                            player_dismissed = batsman
                            dismissal_kind = "caught"
                        elif ball_event.split(" ")[1] == "Caught&Bowled!!":
                            player_dismissed = batsman
                            dismissal_kind = "caught and bowled"
                        elif ball_event.split(" ")[1] == "Stumped!!":
                            player_dismissed = batsman
                            dismissal_kind = "stumped"
                            last_ball_batsman = non_striker
                        elif "Run Out!!" in ball_event:
                            player_dismissed = find_between(ball_event, "out ", " Run Out!!")
                            dismissal_kind = "run out"
                        else:
                            print("Out event not handled !!")
                            #break
                    else:
                        print("Ball event not handled !!")
                        #break

                    balldata = {"match_id":[match_id], "season":[season], "city":[city], "team1":[team1], \
                                "team2":[team2], "winner":[winner], "win_by_runs":[win_by_runs], \
                                "win_by_wickets":[win_by_wickets], "inning":[inning], "venue":[venue], \
                                "over":[over], "ball":[ball], "batsman":[batsman], "bowler":[bowler], \
                                "non_striker":[non_striker], "batting_team":[batting_team], "bowling_team":[bowling_team], \
                                "is_super_over":[is_super_over], "result":[result], "dl_applied":[dl_applied], \
                                "wide_runs":[wide_runs], "bye_runs":[bye_runs], "legbye_runs":[legbye_runs], \
                                "noball_runs":[noball_runs], "penalty_runs":[penalty_runs], "batsman_runs":[batsman_runs], \
                                "extra_runs":[extra_runs], "total_runs":[total_runs], "player_dismissed":[player_dismissed], \
                                "dismissal_kind":[dismissal_kind], "fielder":[fielder], "date":[date], \
                                "toss_winner":[toss_winner], "toss_decision":[toss_decision], "player_of_match": [player_of_match], \
                                "umpire1":[umpire1], "umpire2":[umpire2], "umpire3":[umpire3], \
                               }
                    dfball = pd.DataFrame(balldata)  
                    dfmatch = dfmatch.append(dfball)

                dfballbyball = dfballbyball.append(dfmatch)
        
            if len(ball_id) != 0:
                dfballbyball = dfballbyball[['match_id', 'season', 'city', 'date', 'team1', 'team2', 'toss_winner',
                       'toss_decision', 'result', 'dl_applied', 'winner', 'win_by_runs',
                       'win_by_wickets', 'player_of_match', 'venue', 'umpire1', 'umpire2',
                       'umpire3', 'inning', 'batting_team', 'bowling_team', 'over', 'ball',
                       'batsman', 'non_striker', 'bowler', 'is_super_over', 'wide_runs',
                       'bye_runs', 'legbye_runs', 'noball_runs', 'penalty_runs',
                       'batsman_runs', 'extra_runs', 'total_runs', 'player_dismissed',
                       'dismissal_kind', 'fielder']]
        else:
            print(winners_details[i])

    
    return dfballbyball
```

```python
url = url_syed
res = requests.get(url)
soup1 = BeautifulSoup(res.content,'lxml')
matches = soup1.find_all('a', {'class': 'text-hvr-underline', 'href': True})
venues = soup1.find_all("div", {"class":"text-gray"})
matchid_offset = 1

# Winner details
winners_list = soup1.find_all("a", {"class":"cb-text-link"})
winners_details = []
dl_applied_list = []
for i in range(len(winners_list)):
    win_details = winners_list[i].text
    if win_details.endswith("(DLS method)"):
        win_details = win_details[:-13]
        dl_applied_list.append(1)
    else:
        dl_applied_list.append(0)
    win_details = win_details.split(" won by ")
    winners_details.append(win_details)

if url=="https://www.cricbuzz.com/cricket-series/2717/caribbean-premier-league-2018/matches":
    winners_details[24] = ['St Kitts and Nevis Patriots', '7 wkts']        

dfballbyball = pd.DataFrame()

# Iterate over all matches
matches = matches[:-2]

for i in [4]:#range(len(matches)):

    print(i)
    dfmatch = pd.DataFrame()

    match_href = "http://www.cricbuzz.com" + matches[i]['href']
    venue = venues[i*2+1]
    city = venue.text.split(", ")[1]
    venue = venue.text
    match_id = matchid_offset + i
    season = 2018

    team1 = matches[i].find_all('span')[0].text.split(" vs ")[0]
    team2 = matches[i].find_all('span')[0].text.split(" vs ")[1].split(", ")[0]

    winner = winners_details[i][0]
    is_super_over = 0
    dl_applied = dl_applied_list[i]
    result = "normal"


    ## Details not available through scraping
    date = np.nan
    toss_winner = np.nan
    toss_decision = np.nan
    player_of_match = np.nan
    umpire1 = np.nan
    umpire2 = np.nan
    umpire3 = np.nan


    if winner == team1:
        loser = team2
    else:
        loser = team1


    if (winners_details[i] != ["No result"]) and (winners_details[i] != ["Match tied"]) and (
        winners_details[i] != ["Match abandoned without a ball being bowled(no toss)"]) and (
        winners_details[i] != ["Match abandoned without a ball being bowled (no toss)"]) and (
        winners_details[i] != ["Match abandoned without a ball being bowled (no toss) due to wet outfield"]) and (
        winners_details[i] != ["Match abandoned without toss due to rain"]) and (
        winners_details[i] != ["Match abandoned due to rain (without a ball being bowled)"]) and (
        winners_details[i] != ["Match tied (Islamabad United won the one-over eliminator)"]) and (
        winners_details[i] != ["Match tied (Lahore Qalandars won the super over) "]) and (
        winners_details[i] != ["Match abandoned without a ball being bowled"]) and (
        winners_details[i] != ["Match abandoned without toss due to wet outfield"]) and (
        winners_details[i] != ["Match abandoned without toss due to rain"]) and (
        winners_details[i] != ["Match rescheduled to 01/09/2018 due to bad outfield"]) and (
        winners_details[i] != ["Match tied (Bellary Tuskers won the one-over eliminator)"]) and (
        winners_details[i] != ["Match tied - Punjab won the one-over eliminator"]):
        if "run" in winners_details[i][1]:
        #if winners_details[i][1].endswith=="run" or winners_details[i][1].endswith=="runs":
            win_by_runs = int(winners_details[i][1].split(" ")[0])
            win_by_wickets = 0       
            innings1_batting_team = winner
            innings1_bowling_team = loser
            innings2_batting_team = loser
            innings2_bowling_team = winner
        elif "wkt" in winners_details[i][1]:
            win_by_wickets = int(winners_details[i][1].split(" ")[0])
            win_by_runs = 0
            innings1_batting_team = loser
            innings1_bowling_team = winner
            innings2_batting_team = winner
            innings2_bowling_team = loser
        else:
            print("Error in win detail !!")


        res = requests.get(match_href)
        soup2 = BeautifulSoup(res.content,'lxml')
        ball_id = soup2.find_all("span",{"class":"text-bold","class":"cb-col"})[::-1]
        ball_details = soup2.find_all("p",{"class":"cb-col-90"})[::-1]

        if len(ball_id) != len(ball_details):
            print("Error !!")
            break
        else:
            inning = 1
            last_ball_batsman = ""
            non_striker = ""

            for j in range(len(ball_id)):

                over = int(ball_id[j].text.split(".")[0]) + 1
                ball = int(ball_id[j].text.split(".")[1])

                ## Set innings number
                if (j>20 and over==1 and ball==1):
                    inning = 2
                    last_ball_batsman = ""
                    non_striker = ""

                if inning==1:
                    batting_team = innings1_batting_team
                    bowling_team = innings1_bowling_team
                else:
                    batting_team = innings2_batting_team
                    bowling_team = innings2_bowling_team

                bowler = ball_details[j].text.split(", ")[0].split(" to ")[0]
                batsman = ball_details[j].text.split(", ")[0].split(" to ")[1]

                if batsman != last_ball_batsman:
                    non_striker = last_ball_batsman
                    last_ball_batsman = batsman

                # Check different scenarios for each ball
                ball_event = ball_details[j].text.split(", ")[1]

                wide_runs = 0
                bye_runs = 0
                legbye_runs = 0
                noball_runs = 0
                penalty_runs = 0
                batsman_runs = 0
                extra_runs = 0
                total_runs = 0
                player_dismissed = np.nan
                dismissal_kind = np.nan
                fielder = np.nan

                if ball_event == "wide":
                    wide_runs = 1
                    total_runs = 1
                elif ball_event == "no ball":
                    noball_runs = 1
                    batsman_runs = 0
                    if len(ball_details[j].text.split(", "))>2 and "run" in ball_details[j].text.split(", ")[2]:
                        try:
                            batsman_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                        except:
                            batsman_runs = 0
                    total_runs = noball_runs + batsman_runs                
                elif ball_event=="leg byes" or ball_event=="Leg byes":
                    if "run" in ball_details[j].text.split(", ")[2]:
                        legbye_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                        total_runs = legbye_runs
                elif ball_event == "byes":
                    if "run" in ball_details[j].text.split(", ")[2]:
                        bye_runs = int(ball_details[j].text.split(", ")[2].split(" ")[0])
                        total_runs = bye_runs
                elif "wides" in ball_event:
                    extra_runs = ball_event.split(" ")[0]
                    total_runs = extra_runs
                elif "no run" in ball_event:
                    batsman_runs = 0
                    total_runs = 0
                elif "1 run" in ball_event: # == "1 run" or ball_event=="1 run," or ball_event:
                    batsman_runs = 1
                    total_runs = 1
                elif ball_event == "2 runs":
                    batsman_runs = 2
                    total_runs = 2
                elif ball_event == "3 runs":
                    batsman_runs = 3
                    total_runs = 3
                elif ball_event == "FOUR" or ball_event == "4 runs":
                    batsman_runs = 4
                    total_runs = 4
                elif ball_event == "5 runs":
                    batsman_runs = 5
                    total_runs = 5
                elif ball_event == "SIX":
                    batsman_runs = 6
                    total_runs = 6
                elif ball_event.split(" ")[0] == "out":
                    if ball_event.split(" ")[1] == "Lbw!!":
                        player_dismissed = batsman
                        dismissal_kind = "lbw"
                        last_ball_batsman = non_striker
                    elif ball_event.split(" ")[1] == "Bowled!!":
                        player_dismissed = batsman
                        dismissal_kind = "bowled"
                        last_ball_batsman = non_striker
                    elif ball_event.split(" ")[1] == "Caught":
                        player_dismissed = batsman
                        dismissal_kind = "caught"
                    elif ball_event.split(" ")[1] == "Caught&Bowled!!":
                        player_dismissed = batsman
                        dismissal_kind = "caught and bowled"
                    elif ball_event.split(" ")[1] == "Stumped!!":
                        player_dismissed = batsman
                        dismissal_kind = "stumped"
                        last_ball_batsman = non_striker
                    elif "Run Out!!" in ball_event:
                        player_dismissed = find_between(ball_event, "out ", " Run Out!!")
                        dismissal_kind = "run out"
                    else:
                        print("Out event not handled !!")
                        break
                else:
                    print("Ball event not handled !!")
                    break

                balldata = {"match_id":[match_id], "season":[season], "city":[city], "team1":[team1], \
                            "team2":[team2], "winner":[winner], "win_by_runs":[win_by_runs], \
                            "win_by_wickets":[win_by_wickets], "inning":[inning], "venue":[venue], \
                            "over":[over], "ball":[ball], "batsman":[batsman], "bowler":[bowler], \
                            "non_striker":[non_striker], "batting_team":[batting_team], "bowling_team":[bowling_team], \
                            "is_super_over":[is_super_over], "result":[result], "dl_applied":[dl_applied], \
                            "wide_runs":[wide_runs], "bye_runs":[bye_runs], "legbye_runs":[legbye_runs], \
                            "noball_runs":[noball_runs], "penalty_runs":[penalty_runs], "batsman_runs":[batsman_runs], \
                            "extra_runs":[extra_runs], "total_runs":[total_runs], "player_dismissed":[player_dismissed], \
                            "dismissal_kind":[dismissal_kind], "fielder":[fielder], "date":[date], \
                            "toss_winner":[toss_winner], "toss_decision":[toss_decision], "player_of_match": [player_of_match], \
                            "umpire1":[umpire1], "umpire2":[umpire2], "umpire3":[umpire3], \
                           }
                dfball = pd.DataFrame(balldata)  
                dfmatch = dfmatch.append(dfball)

            dfballbyball = dfballbyball.append(dfmatch)

        if len(ball_id) != 0:
            dfballbyball = dfballbyball[['match_id', 'season', 'city', 'date', 'team1', 'team2', 'toss_winner',
                   'toss_decision', 'result', 'dl_applied', 'winner', 'win_by_runs',
                   'win_by_wickets', 'player_of_match', 'venue', 'umpire1', 'umpire2',
                   'umpire3', 'inning', 'batting_team', 'bowling_team', 'over', 'ball',
                   'batsman', 'non_striker', 'bowler', 'is_super_over', 'wide_runs',
                   'bye_runs', 'legbye_runs', 'noball_runs', 'penalty_runs',
                   'batsman_runs', 'extra_runs', 'total_runs', 'player_dismissed',
                   'dismissal_kind', 'fielder']]
    else:
        print(winners_details[i])
```
